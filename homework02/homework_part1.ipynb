$$\rho(\theta) = (1 + 0.9 \cdot cos (8 \cdot \theta) ) \cdot (1 + 0.1 \cdot cos(24 \cdot \theta)) \cdot (0.9 + 0.05 \cdot cos(200 \cdot \theta)) \cdot (1 + sin(\theta))$$

````python
theta = torch.linspace(-np.pi, np.pi, steps=1000)

# compute rho(theta) as per formula above
rho =   (1. + 0.9 * torch.cos(8. * theta)) \
      * (1. + 0.1 * torch.cos(24. * theta)) \
      * (0.9 + 0.05 * torch.cos(200. * theta)) \
      * (1. + torch.sin(theta))

# Now convert polar (rho, theta) pairs into cartesian (x,y) to plot them.
x = rho * torch.cos(theta)
y = rho * torch.sin(theta)


plt.figure(figsize=[6,6])
plt.fill(x.numpy(), y.numpy(), color='green')
plt.grid()
````

### Task II: the game of life

Now it's time for you to make something more challenging. We'll implement Conway's [Game of Life](http://web.stanford.edu/~cdebs/GameOfLife/) in _pure pytorch_.

![img](https://cdn.tutsplus.com/gamedev/authors/legacy/Stephane%20Beniak/2012/09/11/Preview_Image.png)
If you've skipped the url above out of sloth, here's the game of life:
* You have a 2D grid of cells, where each cell is "alive"(1) or "dead"(0)
* Any living cell that has 2 or 3 neighbors survives, else it dies [0,1 or 4+ neighbors]
* Any cell with exactly 3 neighbors becomes alive (if it was dead)

For this task, you are given a reference numpy implementation that you must convert to pytorch.
_[numpy code inspired by: https://github.com/rougier/numpy-100]_


__Note:__ You can find convolution in `torch.nn.functional.conv2d(Z,filters)`. Note that it has a different input format. 


````python
from scipy.signal import convolve2d

def np_update(Z):
    # Count neighbours with convolution
    filters = np.array([[1,1,1],
                        [1,0,1],
                        [1,1,1]])
    
    N = convolve2d(Z,filters,mode='same')
    
    # Apply rules
    birth = (N==3) & (Z==0)
    survive = ((N==2) | (N==3)) & (Z==1)
    
    Z[:] = birth | survive
    return Z
````

````python
from torch.nn.functional import conv2d
````

````python
from torch.nn.functional import conv2d

def torch_update(Z):
    """
    Implement an update function that does to Z exactly the same as np_update.
    :param Z: torch.FloatTensor of shape [height,width] containing 0s(dead) an 1s(alive)
    :returns: torch.FloatTensor Z after updates.
    
    You can opt to create new tensor or change Z inplace.
    """
    Z = Z.clone()
    Z = Z.resize_(1, 1, Z.shape[0], Z.shape[1])
    Z = Variable(Z)
    filters = Variable(torch.FloatTensor([[[[1,1,1],
                                            [1,0,1],
                                            [1,1,1]]]]))
    N = conv2d(Z, filters, padding=1).data[0, 0]
    Z = Z.data[0, 0]
    
    birth = (N==3) & (Z==0)
    survive = ((N==2) | (N==3)) & (Z==1)
    
    Z[:] = birth | survive
    return Z
````

````python
#initial frame
Z_numpy = np.random.choice([0,1],p=(0.5,0.5),size=(100,100))
Z = torch.from_numpy(Z_numpy).type(torch.FloatTensor)

#your debug polygon :)
Z_new = torch_update(Z.clone())

#tests
Z_reference = np_update(Z_numpy.copy())
assert np.all(Z_new.numpy() == Z_reference), "your pytorch implementation doesn't match np_update. Look into Z and np_update(ZZ) to investigate."
print("Well done!")
````

````python
from notmnist import load_notmnist
X_train, y_train, X_test, y_test = load_notmnist(letters='ABCDEFGHIJ')
X_train, X_test = X_train.reshape([-1, 784]), X_test.reshape([-1, 784])
````

````python
%matplotlib inline
plt.figure(figsize=[12,4])
for i in range(20):
    plt.subplot(2,10,i+1)
    plt.imshow(X_train[i].reshape([28,28]))
    plt.title(str(y_train[i]))
````

````python
from IPython.display import clear_output
from tqdm import trange
from torch import nn
from torch.nn.functional import cross_entropy

def minibatches(X, y, batch_size=64):
    N = len(X)
    indices = np.arange(N)
    np.random.shuffle(indices)
    for start_idx in range(0, N, batch_size):
        yield X[start_idx:start_idx + batch_size], y[start_idx:start_idx + batch_size]

def compute_loss(net, X, y):
    X = Variable(torch.FloatTensor(X))
    y = Variable(torch.LongTensor(y))
    logits = net.forward(X)
    
    return cross_entropy(logits, y).mean()

def compute_score(net, X, y_true):
    net.train(False)
    batch_acc = []
    
    for X_batch, y_batch in minibatches(X, y_true):
        X_batch = Variable(torch.FloatTensor(X_batch))
        logits = net.forward(X_batch)
        y_pred = logits.max(1)[1].data.numpy()
        batch_acc.append(np.sum(y_batch == y_pred))
    
    return np.sum(batch_acc) / len(X)

def train_net(net, X_train, y_train, batch_size=64, n_epochs=50, lr=0.001):
    epochs_loss = []
    
    for epoch in trange(n_epochs):
        opt = torch.optim.Adam(net.parameters(), lr)
        batch_loss = []
        
        net.train(True)
        for X_batch, y_batch in minibatches(X_train, y_train):
            loss = compute_loss(net, X_batch, y_batch)
            loss.backward()
            opt.step()
            opt.zero_grad()
            
            batch_loss.append(loss.data.numpy() * len(X_batch))
        
        epochs_loss.append(np.sum(batch_loss) / len(X_train))
        
        clear_output(True)
        plt.title('Train loss')
        plt.plot(epochs_loss)
        plt.ylabel('loss')
        plt.xlabel('epoch')
        plt.show()
    
    return epochs_loss
````

````python
log_reg = nn.Sequential()
log_reg.add_module('logits', nn.Linear(28 * 28, 10))
````

````python
log = train_net(log_reg, X_train, y_train, n_epochs=10)
````

````python
print('Logistic regression test score:')
print(compute_score(log_reg, X_test, y_test))
````

````python
net = nn.Sequential()
net.add_module('linear1', nn.Linear(28 * 28, 512))
net.add_module('dropout1', nn.Dropout())
net.add_module('linear1_relu', nn.ReLU())
net.add_module('linear2', nn.Linear(512, 512))
net.add_module('dropout2', nn.Dropout())
net.add_module('linear2_relu', nn.ReLU())
net.add_module('logits', nn.Linear(512, 10))
````

````python
log = train_net(net, X_train, y_train, n_epochs=20)
````

````python
print('NN test score:')
print(compute_score(net, X_test, y_test))
````
